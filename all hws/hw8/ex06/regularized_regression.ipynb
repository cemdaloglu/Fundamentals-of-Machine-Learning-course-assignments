# 1 Bias and variance of ridge regression

### Proof of $ E[\hat{\beta_{\tau}}] = S_{\tau} ^ {-1}S\beta^*$

$\frac{\partial{\hat{\beta_{\tau}}}}{\partial{\beta}} = 2[X^T(Y-X\beta) + \tau\beta] = 0$

$X^TY-X^TX\hat{\beta}+\tau\hat{\beta} = 0$

$X^TY = X^TX\hat{\beta}-\tau\hat{\beta}$

$X^TY = [X^TX-\tau I_D]\hat{\beta}$

$\hat{\beta} = [X^TX-\tau I_D]^(-1)X^TY where Y = X\beta^* + \epsilon$

$E[\hat{\beta_{\tau}}] = [X^TX-\tau I_D] ^ {-1} X^T(X\beta^* + \epsilon)$ where $E[\epsilon] = 0$. Therefore;
$E[\hat{\beta_{\tau}}] = [X^TX-\tau I_D] ^ {-1} X^TX\beta^* = S_{\tau} ^ {-1}S\beta^*$

### Proof of $ Cov[\hat{\beta_{\tau}}] = S_{\tau} ^ {-1}SS_{\tau} ^ {-1}\sigma^2$

$Cov[\hat{\beta_{\tau}}] = E[(\hat{\beta_{\tau}} - E(\hat{\beta_{\tau}}))(\hat{\beta_{\tau}} - E(\hat{\beta_{\tau}}))^T]$

$\hat{\beta_{\tau}} - E(\hat{\beta_{\tau}}) = S_{\tau} ^ {-1}X^TY - (S_{\tau} ^ {-1}X^TXX^{-1}(Y-\epsilon)$ since $\beta^* = X^{-1}(Y - \epsilon)$ Therefore,

$Cov[\hat{\beta_{\tau}}] = E[(S_{\tau} ^ {-1}X^T\epsilon)(S_{\tau} ^ {-1}X^T\epsilon)^T] = E[(S_{\tau} ^ {-1}X^T\epsilon)(\epsilon XS_{\tau} ^ {-1})]$ since $S_{\tau} ^ {-1} = (S_{\tau} ^ {-1})^T$

Now the equation becomes,

$Cov[\hat{\beta_{\tau}}] = E[\epsilon^2 S_{\tau} ^ {-1} S S_{\tau} ^ {-1}] = E[\epsilon^2] S_{\tau} ^ {-1} S S_{\tau} ^ {-1}$ We know that $var[\epsilon] = E[\epsilon^2] - E[\epsilon]^2 = \sigma^2$

This concludes the proof that,

$Cov[\hat{\beta_{\tau}}] = \sigma^2S_{\tau} ^ {-1} S S_{\tau} ^ {-1}$ 

# 2 Denoising of a CT image

```python
import numpy as np   
import matplotlib.pyplot as plt  
from scipy.sparse import coo_matrix
from scipy.sparse import csc_matrix
from scipy.sparse import linalg
from scipy.sparse import dok_matrix, coo_matrix, vstack
from scipy.ndimage.filters import gaussian_filter
from sklearn.utils import shuffle
from sklearn.datasets import load_digits
from sklearn import model_selection
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
def construct_X(M, alphas, Np = None, tau=0):
    if Np is None:
        Np = int(np.ceil(np.sqrt(2) * M))
        if Np % 2 == 0:
            Np = Np + 1
    
    D = M * M
    No = len(alphas)
    N = No * Np
    
    alphas_rad = np.radians(alphas)
    j = np.arange(D)
    
    M2 = (M-1) / 2
    C = np.mgrid[-M2:M-M2,-M2:M-M2]
    C = C.swapaxes(1,2)
    C = C.reshape(2,D)
    
    i = np.zeros(D * 2 * len(alphas_rad), dtype = np.int32)
    j_ab = np.zeros(i.shape, dtype = np.int32)
    weights = np.zeros(i.shape, dtype = np.float32)
    
    current_orient = np.zeros((len(alphas), 2))
    for index, alps in enumerate(alphas_rad):
        current_orient[index, :] = np.array([np.cos(alps), -np.sin(alps)])

    s0 = Np // 2
    p = current_orient @ C + s0
    i_p = np.floor(p)
    mat_ind = 2 * D
    
    for k in range(len(alphas_rad)):
        i[k*mat_ind:(k+1)*mat_ind] = np.concatenate(((i_p[k] + k*Np), (i_p[k]+1 + k*Np)), axis = None)
        
        j_ab[k*mat_ind:(k+1)*mat_ind] = np.concatenate((j, j), axis = None)
          
        weights[k*mat_ind : (k+1)*mat_ind - D] = (i_p[k]+1) - p[k]
        weights[k*mat_ind + D : (k+1)*mat_ind] = p[k] - i_p[k]
    
    X = coo_matrix((weights, (i, j_ab)), shape = (N, D), dtype = np.float32)
    
    if tau > 0:
        ridge_part = np.diag(np.full(D, np.sqrt(tau)))
        X = vstack((X, ridge_part))
    
    return X
```

```python
y = np.load('hs_tomography/y_195.npy')
alphas = np.load('hs_tomography/alphas_195.npy')

M = 195
Np = 275
proj_no = 64
D = M * M

y = np.reshape(y, (alphas.shape[0], Np))
y_sub = y[:proj_no, :]
y_sub = np.reshape(y_sub, (y_sub.shape[0] * Np))
y_sub_tau = np.hstack((y_sub, np.zeros(D)))

tau = [0,1,10,100,1000,10000]
plot_arr = []

for i in range(len(tau)):
    X = construct_X(M, alphas[:proj_no], Np, tau=tau[i]).tocsc()
    if tau[i] == 0:
        beta = linalg.lsqr(X, y_sub, atol = 1e-5, btol = 1e-5)[0].reshape(M, M)
    else:
        beta = linalg.lsqr(X, y_sub_tau, atol = 1e-5, btol = 1e-5)[0].reshape(M, M)
    plot_arr.append(beta)
```

```python
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
plt.gray()
for x_dim in range(3):
    for y_dim in range(2):
        axes[x_dim, y_dim].imshow(plot_arr[x_dim*2 + y_dim], aspect = 'equal')
        axes[x_dim, y_dim].set_title('τ = {}'.format(tau[x_dim*2 + y_dim]));
        axes[x_dim, y_dim].axis('off')
plt.show()
```

```python
X = construct_X(M, alphas[:proj_no], Np, tau=0).tocsc()
beta = linalg.lsqr(X, y_sub, atol = 1e-5, btol = 1e-5)[0].reshape(M, M)

sigma = [0,1,2,3,5,7]
fig, axes = plt.subplots(3, 2, figsize = (18, 18))

for x_dim in range(3):
    for y_dim in range(2):
        beta_gauss = gaussian_filter(beta, sigma[x_dim*2 + y_dim])
        axes[x_dim, y_dim].imshow(beta_gauss, aspect = 'equal')
        axes[x_dim, y_dim].set_title('σ = {}'.format(sigma[x_dim*2 + y_dim]));
        axes[x_dim, y_dim].axis('off')
plt.show()
```

Gaussian filtering averages everything so we lose the details as sigma increases. However, we can still distinguish the edges. On the ridge regression results, details can be seen but as tau increases we lose the boundaries of the objects in the image.

## 3.1 Implement Orthogonal Matching Pursuit

```python
def omp_regression(X, y, T):
    # initialization
    D = X.shape[1]
    beta_hat = np.zeros((D, T))
    A = []
    B = np.arange(D).tolist()
    r_t = y
    
    # iteration
    for t in range(T):
        j_t = np.argmax(np.abs([X[:, j].T @ r_t for j in B]))
        B.remove(B[j_t])
        A.append(j_t)
        X_t = X[:, A]
        beta_hat[A, t] = linalg.lsqr(X_t, y)[0]
        r_t = y - X_t @ beta_hat[A, t]
    return beta_hat
```

```python
def predict_error(test, beta, label):
    prediction = test @ beta
    prediction[prediction < 0] = -1
    prediction[prediction > 0] = 1
    return 100 * (1 - np.count_nonzero(prediction == label) / len(prediction))
```

```python
T = 12
beta_std = omp_regression(x_train_1_7_std, y_train_1_7, T)
beta = omp_regression(x_train_1_7, y_train_1_7, T)

error_arr = np.zeros((4, T))
for t in range(T):
    error_arr[0,t] = predict_error(x_train_1_7_std, beta_std[:,t], y_train_1_7)
    error_arr[1,t] = predict_error(x_test_1_7_std, beta_std[:,t], y_test_1_7)
    error_arr[2,t] = predict_error(x_train_1_7, beta[:,t], y_train_1_7)
    error_arr[3,t] = predict_error(x_test_1_7, beta[:,t], y_test_1_7)

x_axis = np.arange(T) + 1
plt.xlabel('t')
plt.ylabel('Error')
plt.title('OMP Error Rates')
plt.plot(x_axis, error_arr[0,:], 'r', label='train_std')
plt.plot(x_axis, error_arr[1,:], label='test_std')
plt.plot(x_axis, error_arr[2,:], 'g', label='train')
plt.plot(x_axis, error_arr[3,:], label='test')
plt.xticks(x_axis)
plt.legend()
plt.show()
```

Standardization only helped for the small t values. I think it is not necessary to standardize after t > 5.

```python
def preprocess(k, data, target):
    data, target = shuffle(data, target, random_state=0)
    
    data_pos = data[target == k]
    data_neg = data[target != k]
    target_pos = target[target == k]
    target_neg = target[target != k]
    
    data_neg = data_neg[:data_pos.shape[0],:]
    target_neg = target_neg[:data_pos.shape[0]]
   
    data = np.concatenate((data_pos, data_neg))
    target = np.concatenate((target_pos, target_neg))
    
    target[target != k] = -1
    target[target == k] = 1
    
    x_train, y_train = shuffle(data, target, random_state=0)
    return x_train, y_train
```
